```python
def getAllPlayersCareers(allPlayers):
  print('--------------------------------')
  # with open(txtFilePath, 'r') as f:
  #   file = f.read()
  #   if(len(file) != 0):
  #     allPlayers = eval(file)
  print(f"Total Players {len(allPlayers)}")
  careers = pd.DataFrame()
  count = 1
  for player in allPlayers:

    playerName = player['full_name']
    playerId = player['id']
    print(f"{count} : Pulling data for Player: {playerName}, ID: {playerId}")
    try:
      career = playercareerstats.PlayerCareerStats(player_id=playerId) 
      career = career.get_data_frames()[0]
      career['player_name'] = playerName
      # allPlayers[:] = [d for d in allPlayers if d.get('id') != playerId]
      # with open(txtFilePath, 'w') as f:
      #   f.write(str(allPlayers))
      updateCSV(career)
      count = count + 1
    except:
      print(f"Pulling data FAILED for Player: {playerName}, ID: {playerId}")
  

  return careers
```
