# Network Measures

```python
# Import the necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context("poster")
plt.style.use('seaborn')

import pandas as pd
import numpy as np
```

```python
with_lebron = pd.read_csv('../Results/04_Network Data Dumps/withLebron.csv')
```

```python
without_lebron = pd.read_csv('../Results/04_Network Data Dumps/withoutLebron.csv')
```

```python
with_lebron.head()
```

## With Lebron Measures

```python
with_lebron = with_lebron.sort_values('Weighted Degree',ascending=False)
with_lebron.head(10)

with_lebron['color'] = '#143D52'
# Set the color for specific players
with_lebron.loc[with_lebron['Id'] == 'LeBron James', 'color'] = '#00BFFF'

# Plot the bar chart
plt.figure(figsize=(15,8))
sns.barplot(x=with_lebron['Id'].head(20), y=with_lebron['Weighted Degree'].head(20),palette=with_lebron['color'])
plt.xticks(rotation=60,fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Players', fontsize=20)
plt.ylabel('Weighted Degree', fontsize=20)
plt.title("Network Measures - Weighted Degree", fontsize=25)
plt.show()

image_format = 'svg' # e.g .png, .svg, etc.
image_name = 'With Lebron Weighted Degree.svg'
image_path = '../Figures/05_Network_Measures/'

plt.savefig(image_path+image_name, format=image_format, dpi=1200)
```
